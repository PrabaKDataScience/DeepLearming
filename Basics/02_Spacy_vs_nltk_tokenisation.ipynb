```python
import spacy
```

# Sentence tokenisation

```python
nlp = spacy.load("en_core_web_sm")

test_str = "Mr. Praba want to study in abroad. He is interested in moving to countries like Switzerland , India , Singapore etc. He is going there soon . "
doc = nlp(test_str)

for sentence in doc.sents:
  print(sentence)
```

# Word tokenisation 

```python
doc = nlp("Mr. Praba want to study in abroad. He is interested in moving to countries like Switzerland , India , Singapore etc. He is going there soon . ")

for sentence in doc.sents:
  for word in sentence:
    print(word)
```
