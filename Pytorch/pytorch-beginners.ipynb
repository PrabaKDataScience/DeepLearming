```python
if torch.cuda.is_available():
    print('CUDA found')
    # Creating tensor in GPU
    device = torch.device('CUDA')
    x = torch.ones(5,device=device)
    
    #creating a tensor in CPU
    y = torch.ones(5)
    #moving it to GPU
    y =y.to(device)
    
    # Now the arithmetic will un in GPU  FASTER 
    z=x+y
    
    
    # If you want a numpy version , you have to move the varibale to CPU , Numpy cant support GPU
    res = z.to('cpu')
```
